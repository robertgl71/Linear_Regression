```python
#Exercici Q2
import numpy as np

MEDV_List = np.loadtxt("housing.data",usecols=(13,))
counter = 0
suma_MEDV = 0
suma_MSE = 0

for i in MEDV_List:
    counter += 1
    suma_MEDV += i

mitja = suma_MEDV / counter

for j in MEDV_List:
    suma_MSE += (j-mitja)**2
MSE = suma_MSE/counter

print "MEDV: " + str(mitja)
print "MSE: " + str(MSE)
```

```python
#Exercici Q4
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model

dataX= np.loadtxt("housing.data",usecols=range(0,12))
dataY = np.loadtxt("housing.data",usecols=(13,))

fullSizeY = dataY.size
midSizeY=dataY.size/2

# Split the data into training/testing sets
dataX_train = dataX[0:midSizeY]
dataX_test = dataX[midSizeY:fullSizeY]
     
# Split the targets into training/testing sets
dataY_train = dataY[0:midSizeY]
dataY_test = dataY[midSizeY:fullSizeY]

# Create linear regression object
regr1 = linear_model.LinearRegression()
regr2 = linear_model.LinearRegression()
     
#Train the model using the training sets
regr1.fit(dataX_train, dataY_train)
regr2.fit(dataX_test , dataY_test)
print "With all features"
print "\n"
print("MSE Training Data: %.2f"
          % np.mean((regr1.predict(dataX_train) - dataY_train) ** 2))
print("MSE Test Data: %.2f"
          % np.mean((regr2.predict(dataX_test) - dataY_test) ** 2))

print ("R^2 Train: %.2f"
           % regr1.score(dataX_train, dataY_train))
print ("R^2 Test: %.2f"
           % regr2.score(dataX_test, dataY_test))
print "\n"
print "Without feature CHAS"
print "\n"
dataX= np.loadtxt("housing.data",usecols=range(0,2)+range(4,12))
dataY = np.loadtxt("housing.data",usecols=(13,))

fullSizeY = dataY.size
midSizeY=dataY.size/2

# Split the data into training/testing sets
dataX_train = dataX[0:midSizeY]
dataX_test = dataX[midSizeY:fullSizeY]
     
# Split the targets into training/testing sets
dataY_train = dataY[0:midSizeY]
dataY_test = dataY[midSizeY:fullSizeY]

# Create linear regression object
regr3 = linear_model.LinearRegression()
regr4 = linear_model.LinearRegression()
     
#Train the model using the training sets
regr3.fit(dataX_train, dataY_train)
regr4.fit(dataX_test, dataY_test)
# 
print("MSE Training Data: %.2f"
          % np.mean((regr3.predict(dataX_train) - dataY_train) ** 2))
print("MSE Test Data: %.2f"
          % np.mean((regr4.predict(dataX_test) - dataY_test) ** 2))

print ("R^2 Train: %.2f"
           % regr3.score(dataX_train, dataY_train))
print ("R^2 Test: %.2f"
           % regr4.score(dataX_test, dataY_test))
print "\nWithout the CHAS feature, the global model go worst."
```

```python
#Exercici Q5
import matplotlib.pyplot as plt
import numpy as np
from numpy import polyfit, poly1d
import warnings
warnings.filterwarnings("ignore")

listAtrib = ['CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT']

dataY = np.loadtxt("housing.data",usecols=(13,))

fullSize = dataY.size
midSize = dataY.size/2

# Split the targets into training/testing sets
dataY_train = dataY[0:midSize]
dataY_test = dataY[midSize:fullSize]

for i in range(0, 13):
    dataX= np.loadtxt("housing.data",usecols=(i,))
    
    # Split the data into training/testing sets
    dataX_train = dataX[0:midSize]
    dataX_test = dataX[midSize:fullSize]
    
    for j in range(1,5):
        #Calcula els coeficients de les x del polinomi
        coefs = polyfit(dataX_train, dataY_train, j)
    
        #Crea el polinomi amb els coeficients creats
        p = poly1d(coefs)
    
        #Ordena el valors de x
        x = sorted(dataX_test)
        
        #Array de sortida amb els valor y per els diferents valors de x
        y = p(x)
        
        plt.figure(i+1,figsize=(8,4))
        plt.suptitle(listAtrib[i])
        plt.subplot(2,2,j)
        plt.title('Grau ' + str(j))
        plt.scatter(dataX_test, dataY_test,  color='black')
        plt.plot(x,y, color='blue',linewidth=3)
        plt.xticks(())
plt.show()
print "Com es pot observar a els gràfics, conforme aumenta el grau del polinomi,"
print "tendeix a no generalitzxar tan be, fins que es produeix overfitting."
```
